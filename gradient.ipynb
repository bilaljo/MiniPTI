```python
from sympy import *
init_printing()
```

```python
x,y = symbols("x y")
I_1i, I_2i, I_3i = symbols("I_1i I_2i I_3i")
I_1j, I_2j, I_3j = symbols("I_1j I_2j I_3j")
i, n, j = symbols("i n j")
```

```python
phi_i = (I_1i + I_2i * cos(x) + I_3i * cos(y)) ** 2 + (I_2i * sin(x) + I_3i * sin(y)) ** 2
phi_j = (I_1j + I_2j * cos(x) + I_3j * cos(y)) ** 2 + (I_2j * sin(x) + I_3j * sin(y)) ** 2
mean = 1/n * Sum(phi_j, (j, 1, n))
mean
circle = Sum((phi_i - mean) ** 2, (i, 1, n))
circle
```

```python
Eq(S("phi_i"), (I_1i + I_2i * cos(x) + I_3i * cos(y)) ** 2 + (I_2i * sin(x) + I_3i * sin(y)) ** 2)
```

```python
phi_i_symbol = symbols("phi_i")
phi_j_symbol = symbols("phi_j")
circle = Sum((phi_i_symbol - 1/n * Sum(phi_j_symbol, (j, 1, n))) ** 2, (i, 1, n))
circle
```

```python
psi_x_j = symbols("psi_x_j")
psi_x_i= symbols("psi_x_i")
Eq(psi_x_i, diff(phi_i, x))
```

```python
psi_y_j = symbols("psi_y_j")
psi_y_i= symbols("psi_y_i")
Eq(psi_y_i, diff(phi_i, y))
```

```python
x_component = (psi_x_i - 1/n * Sum(psi_x_j, (j, 1, n)))
y_component =  (psi_y_i - 1/n * Sum(psi_y_j, (j, 1, n)))
gradient = 2 *Sum( (phi_i_symbol - 1/n * Sum(phi_j_symbol, (j, 1, n))) * Matrix([x_component, y_component]), (i, 1, n))
gradient
```
