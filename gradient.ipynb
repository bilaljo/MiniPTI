```python
from sympy import *
init_printing()
```

```python
x,y = symbols("x y")
I_1i, I_2i, I_3i = symbols("I_1i I_2i I_3i")
I_1j, I_2j, I_3j = symbols("I_1j I_2j I_3j")
i, n, j = symbols("i n j")
```

```python
phi_i = (I_1i + I_2i * cos(x) + I_3i * cos(y)) ** 2 + (I_2i * sin(x) + I_3i * sin(y)) ** 2
phi_j = (I_1j + I_2j * cos(x) + I_3j * cos(y)) ** 2 + (I_2j * sin(x) + I_3j * sin(y)) ** 2
mean = 1/n * Sum(phi_j, (j, 1, n))
mean
circle = Sum((phi_i - mean) ** 2, (i, 1, n))
circle
```

```python
phi_i_symbol = symbols("phi_i")
phi_j_symbol = symbols("phi_j")
circle = Sum((phi_i_symbol - 1/n * Sum(phi_j_symbol, (j, 1, n))) ** 2, (i, 1, n))
circle
```

```python
phi_diff_x = diff(phi_i, x)
phi_diff_x
```

```python
psi_x = symbols("psi_x_i")
psi_xj = symbols("psi_xj")
Sum(psi_x - 1/n * Sum(psi_xj, (j, 1, n)), (i, 1, n))
```

```python
partial_y = diff(phi_i, y)
partial_y
```

```python
psi_y = symbols("psi_yi")
psi_yj = symbols("psi_yj")
Sum(psi_y - 1/n * Sum(psi_yj, (j, 1, n)), (i, 1, n))
```

```python
x_component = 2 * Sum(psi_xj - 1/n * Sum(psi_xj, (j, 1, n)), (i, 1, n)) * Sum((phi_i_symbol - 1/n * Sum(phi_j_symbol, (j, 1, n))), (i, 1, n))
y_component =  2 * Sum(psi_yj - 1/n * Sum(psi_yj, (j, 1, n)), (i, 1, n)) * Sum((phi_i_symbol - 1/n * Sum(phi_j_symbol, (j, 1, n))), (i, 1, n))
gradient = Matrix([x_component, y_component])
gradient
```
