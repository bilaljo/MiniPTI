```python
from sympy import *
import sys
sys.setrecursionlimit(1500000)
init_printing()
```

```python
phi_1, phi_2 = symbols("varphi_1 varphi_2")
I_1, I_2, I_3 = IndexedBase("I_1", shape=400), IndexedBase("I_2", shape=400), IndexedBase("I_3", shape=400)
i, n, j, k = symbols("i n j k")
phi_i_symbol = Function("xi_i")(phi_1, phi_2)
```

```python
phi_i = (I_1[i] + I_2[i] * cos(phi_1) + I_3[i] * cos(phi_2)) ** 2 + (I_2[i] * sin(phi_1) + I_3[i] * sin(phi_2)) ** 2
```

```python
mean = 1/n * Sum(phi_i, (i, 1, n))
mean
```

```python
circle = Sum((phi_i - mean) ** 2, (i, 1, n))
circle
```

```python
Eq(S("xi_i(x,y)"), phi_i)
```

```python
circle_symbolic = Sum((phi_i_symbol - 1/n * Sum(phi_i_symbol, (i, 1, n))) ** 2, (i, 1, n))
circle_symbolic
```

```python
psi_x = diff(phi_i, phi_1)
psi_x_symbol = Function("psi_x_i")(phi_1, phi_2)
Eq(psi_x_symbol, diff(phi_i, phi_1))
```

```python
psi_y = diff(phi_i, phi_2)
psi_y_symbol =  Function("psi_y_i")(phi_1, phi_2)
Eq(psi_y_symbol, diff(phi_i, phi_2))
```

```python
x_component = (psi_x_symbol - 1/n * Sum(psi_x_symbol, (i, 1, n)))
y_component =  (psi_y_symbol - 1/n * Sum(psi_y_symbol, (i, 1, n)))
gradient = Matrix([x_component, y_component])
gradient = 2 *Sum((phi_i_symbol - 1/n * Sum(phi_i_symbol, (i, 1, n))) * Matrix([x_component, y_component]), (i, 1, n))
gradient
```
